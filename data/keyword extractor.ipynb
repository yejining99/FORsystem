```python
import yake
import pandas as pd
```

```python
data = pd.read_csv('./articles.csv')
```

```python
data
```

```python
desc_list=list(data['detail_desc'])
```

```python

key_list=[]
for i in desc_list:
    key_list.append(key_ex(i))
```

```python
def key_ex(text,num=1):
    language = "en"
    max_ngram_size = 1
    deduplication_thresold = 0.9
    deduplication_algo = 'seqm'
    windowSize = 1
    numOfKeywords = num
    
    custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
    keywords = custom_kw_extractor.extract_keywords(text)

    temp=''
    for kw in keywords:
        return kw[0]
```

```python
key_list
```

```python
import csv

with open('key_list.csv','w' ,encoding='UTF-8') as file :

    write = csv.writer(file)
    for i in key_list:
        write.writerow([i])
```
